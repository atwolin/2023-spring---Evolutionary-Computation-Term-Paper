```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np

cur_path = os.getcwd()
print(cur_path)

plt.figure(figsize=(10, 6))

# 保存所有运行结果的列表
word2vec = []
glove = []
fasttext = []

# 读取 word2vec 的数据
for run in range(1, 2):  # 假设文件命名 run=1
    file_path = f"{cur_path}/results/-algo*gpab*-e*word2vec*-n*10*-p*100*-pc*1.0*-pm*0.1*-g*100*-c*0*-eval*1000*-run*{run}.csv"
    data = pd.read_csv(file_path, usecols=['eval_count', 'avg'])  # 只读取 eval_count 和 avg 列
    word2vec.append(data)

# 读取 glove 的数据
for run in range(1, 2):  # 假设文件命名 run=1
    file_path = f"{cur_path}/results/-algo*gpab*-e*glove*-n*10*-p*100*-pc*1.0*-pm*0.1*-g*100*-c*0*-eval*1000*-run*{run}.csv"
    data = pd.read_csv(file_path, usecols=['eval_count', 'avg'])  # 只读取 eval_count 和 avg 列
    glove.append(data)

# 读取 fasttext 的数据
for run in range(1, 2):  # 假设文件命名 run=1
    file_path = f"{cur_path}/results/-algo*gpab*-e*fasttext*-n*10*-p*100*-pc*1.0*-pm*0.1*-g*100*-c*0*-eval*1000*-run*{run}.csv"
    data = pd.read_csv(file_path, usecols=['eval_count', 'avg'])  # 只读取 eval_count 和 avg 列
    fasttext.append(data)

# 将所有运行的数据转换为 np.array 并计算平均值
def compute_avg(data_list):
    # 确保所有运行的数据长度一致
    min_length = min(len(run_data) for run_data in data_list)
    eval_counts = data_list[0]['eval_count'].values[:min_length]
    avg_values = np.array([run_data['avg'].values[:min_length] for run_data in data_list])
    avg_mean = avg_values.mean(axis=0)
    return eval_counts, avg_mean

# 计算每个 eval_count 的平均值
word2vec_eval_counts, avg_word2vec = compute_avg(word2vec)
glove_eval_counts, avg_glove = compute_avg(glove)
fasttext_eval_counts, avg_fasttext = compute_avg(fasttext)

# 绘制平均值的线条
plt.plot(word2vec_eval_counts, avg_word2vec, linestyle='-', linewidth=1, alpha=0.7, label='word2vec')
plt.plot(glove_eval_counts, avg_glove, linestyle='-', linewidth=1, alpha=0.7, label='glove')
plt.plot(fasttext_eval_counts, avg_fasttext, linestyle='-', linewidth=1, alpha=0.7, label='fasttext')

# 绘制图表
plt.xlabel('Eval Count')
plt.ylabel('Avg')
plt.title('Anytime Behavior on dim 20 with different embedding type')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
